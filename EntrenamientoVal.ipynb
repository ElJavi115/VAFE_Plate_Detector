```python
DATASET_YAML = "/content/drive/MyDrive/dataset/placas/data.yaml"
```

```python
import yaml

with open(DATASET_YAML, "r") as f:
    cfg = yaml.safe_load(f)

cfg
```

```python
model = YOLO("yolov8s.pt")

model.train(
    data=DATASET_YAML,
    epochs=120,
    imgsz=640,
    batch=16,
    patience=30,
    lr0=0.001,
    name="placas_v1",
)
```

```python
model.val(data=DATASET_YAML)
```
